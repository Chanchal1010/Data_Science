```python
import random

def estimate_area_of_figure(num_points, bounding_box_width, bounding_box_height, figure_function):
    points_inside_figure = 0
    
    for _ in range(num_points):
        x = random.uniform(0, bounding_box_width)
        y = random.uniform(0, bounding_box_height)
        
        if figure_function(x, y):
            points_inside_figure += 1
    
    bounding_box_area = bounding_box_width * bounding_box_height
    figure_area = (points_inside_figure / num_points) * bounding_box_area
    
    return figure_area

# Example: Estimate the area of a circle with radius 1 enclosed in a square with side length 2
def is_inside_circle(x, y):
    return x**2 + y**2 <= 1

estimated_area = estimate_area_of_figure(1000000, 2, 2, is_inside_circle)
print("Estimated area of the circle:", estimated_area)
```

```python
import random
import matplotlib.pyplot as plt

def estimate_area_of_figure(num_points, bounding_box_width, bounding_box_height, figure_function):
    points_inside_figure = 0
    points_outside_figure_x = []
    points_outside_figure_y = []
    points_inside_figure_x = []
    points_inside_figure_y = []
    
    for _ in range(num_points):
        x = random.uniform(0, bounding_box_width)
        y = random.uniform(0, bounding_box_height)
        
        if figure_function(x, y):
            points_inside_figure += 1
            points_inside_figure_x.append(x)
            points_inside_figure_y.append(y)
        else:
            points_outside_figure_x.append(x)
            points_outside_figure_y.append(y)
    
    bounding_box_area = bounding_box_width * bounding_box_height
    figure_area = (points_inside_figure / num_points) * bounding_box_area
    
    return figure_area, points_outside_figure_x, points_outside_figure_y, points_inside_figure_x, points_inside_figure_y

# Example: Estimate the area of a circle with radius 1 enclosed in a square with side length 2
def is_inside_circle(x, y):
    return x**2 + y**2 <= 1

num_points = 10000
bounding_box_width = 2
bounding_box_height = 2

estimated_area, points_outside_figure_x, points_outside_figure_y, points_inside_figure_x, points_inside_figure_y 
= estimate_area_of_figure(num_points, bounding_box_width, bounding_box_height, is_inside_circle)

# Plotting
plt.figure(figsize=(8, 8))
plt.plot(points_outside_figure_x, points_outside_figure_y, 'r.', label='Outside figure')
plt.plot(points_inside_figure_x, points_inside_figure_y, 'b.', label='Inside figure')
circle = plt.Circle((0, 0), 1, color='k', fill=False, linestyle='--', label='Circle')
plt.gca().add_patch(circle)
plt.xlim(0, bounding_box_width)
plt.ylim(0, bounding_box_height)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Monte Carlo estimation of circle area')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print("Estimated area of the circle:", estimated_area)
```
